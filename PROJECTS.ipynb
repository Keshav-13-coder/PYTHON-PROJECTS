```python
#PROJECT2 ;"STOPWATCH" by_KESHAV.V
import tkinter

root = tkinter.Tk()

root.title('STOPWATCH')
root.geometry('600x300')

time=  tkinter.Label(root, text='00:00:00', width=30)
time.grid(row=0, column=1)

s  = 1
m =  0
h =  0
stop_logic=False
  
def infinite_logic(s,m,h):
    time.config(text=f'0{h}:0{m}:0{s}')
    if s>=10:
     time.config(text=f'0{h}:0{m}:{s}')
    print(s,'sec')
    s=s+1
    if m>=10:
     time.config(text=f'0{h}:{m}:{s}')
     if m>=60:
        m=0
        h = h+1
        time.config(text=f'{h}:0{m}:0{s}')
     if h>=10:
        time.config(text=f'{h}:{m}:{s}')
    str(print('\tmin',m))
    if s>=60:
       s=0
       m=m+1
    global timings
    timings=root.after(1000,infinite_logic,s,m,h)


def fun_start():
     timings=root.after(1000,infinite_logic,s,m,h)
     print('STARTED')

btn2=tkinter.Button(root,text='START',command=lambda sec=s,minute=m,hour=h:infinite_logic(sec,minute,hour));print('PRESS START BUTTON')
btn2.grid(row=1,column=2)

label=  tkinter.Label(root, text='STOP WATCH\n\tby_KESHAV.V', width=17)
label.grid(row=1, column=4)

btn2=tkinter.Button(root,text='Reset & start',command=fun_start)
btn2.grid(row=3,column=1)


def fun_stop():
    root.after_cancel(timings)
    print('STOP BUTTON WAS PRESSED')


def fun_reset():
     root.after_cancel(timings)
     time.config(text='00:00:00')
     global s,m
     s=1
     m=0
     print('RESET BUTTON WAS PRESSED')

def rest_1():
   l=tkinter.Label(root, text='THANK YOUüëã', width=30)
   l.grid(row=6, column=0)
   print('FEEDBACK;GOOD')

def rest_2():
   l=tkinter.Label(root, text='OK,thank youüòÉ', width=30)
   l.grid(row=6, column=1)
   print('FEEDBACK;OK')

def rest_3():
   l=tkinter.Label(root, text='I WILL IMPROVE ITüëç', width=30)
   l.grid(row=6, column=2)
   print('FEEDBACK;BAD')




tkinter.Button(root, text='GOOD',command=rest_1).grid(row=5, column=0)               

tkinter.Button(root, text='OK',command=rest_2).grid(row=5, column=1)  

tkinter.Button(root, text='BAD',command=rest_3).grid(row=5, column=2)


btn1=tkinter.Button(root,text='RESET',command=fun_reset)
btn1.grid(row=1,column=1)


btn=tkinter.Button(root,text='STOP',command=fun_stop)
btn.grid(row=1,column=0)


root.mainloop() 
```
